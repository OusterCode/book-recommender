```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

from dotenv import load_dotenv

import pandas as pd
```

```python
load_dotenv()
```

```python
books = pd.read_csv("books_cleaned.csv")
```

```python
books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_descrition.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False
                                   )
```

```python
raw_documents = TextLoader("tagged_descrition.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding = OpenAIEmbeddings()
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendation(
    query: str,
    top_k: int =10,
    
)-> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)
        
    books_list = []
    
    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]
    
    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendation("A book to teach children about nature")
```
